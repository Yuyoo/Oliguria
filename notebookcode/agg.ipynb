## 将24小时内的数据进行简单的聚合

```python
import pandas as pd
ce0=pd.read_csv('../data/first24chartevents.csv')
le0=pd.read_csv('../data/first24labevents.csv')
ce1=pd.read_csv('../data/chartevents.csv')
le1=pd.read_csv('../data/labevents.csv')
```

```python
ce0_grouped = ce0.drop(['icustay_id', 'charthour'], axis=1).groupby(ce0['icustay_id'])
```

```python
ce0_grouped
```

```python
# 按icustay_id聚合
ce0_grouped = ce0.drop(['icustay_id', 'charthour'], axis=1).groupby(ce0['icustay_id'])
ce1_grouped = ce1.drop(['icustay_id', 'charthour'], axis=1).groupby(ce1['icustay_id'])
le0_grouped = le0.drop(['icustay_id', 'charthour'], axis=1).groupby(le0['icustay_id'])
le1_grouped = le1.drop(['icustay_id', 'charthour'], axis=1).groupby(le1['icustay_id'])
```

```python
# 对ce0中的指标取最大最小值
ce0_max = ce0_grouped.max().rename(columns=lambda x: x + '_max')
ce0_min = ce0_grouped.min().rename(columns=lambda x: x + '_min')
```

```python
ce0_max
```

```python
ce0
```

```python
ce0.dropna(axis=0,how='all')
```

```python
# 连接最大值最小值
ce0_feat=pd.merge(ce0_max,ce0_min,left_index=True,right_index=True)
```

```python
ce0_feat
```

```python
# labevent0的特征
le0_max = le0_grouped.max().rename(columns=lambda x: x + '_max')
le0_min = le0_grouped.min().rename(columns=lambda x: x + '_min')
# 连接
le0_feat=pd.merge(le0_max,le0_min,left_index=True,right_index=True)
```

```python
le0_feat
```

```python
# 前24h人群的特征
feat0 = pd.merge(ce0_feat,le0_feat,left_index=True,right_index=True,how='inner')
```

```python
feat0
```

## 正样本
正样本取少尿患者少尿发生前24h的各项指标的最大最小值

```python
# chartevent1的特征——最大最小值
ce1_max = ce1_grouped.max().rename(columns=lambda x: x + '_max')
ce1_min = ce1_grouped.min().rename(columns=lambda x: x + '_min')
# 连接最大最小值
ce1_feat=pd.merge(ce1_max,ce1_min,left_index=True,right_index=True)
```

```python
# labevent1的特征——最大最小值
le1_max = le1_grouped.max().rename(columns=lambda x: x + '_max')
le1_min = le1_grouped.min().rename(columns=lambda x: x + '_min')
# 连接最大最小值
le1_feat=pd.merge(le1_max,le1_min,left_index=True,right_index=True)
```

```python
feat1=pd.merge(ce1_feat,le1_feat,left_index=True,right_index=True,how='inner')
```

```python
feat1
```

```python
# 增加标志列
feat0['label']=0
feat1['label']=1
```

## 合并数据集，并生成新id
由于本数据集使用同一少尿患者的不同阶段构造。因此正样本的icustayid与负样本的icustayid存在大量重复值，不便于索引。因此需要生成新的唯一id作为索引。

```python
# 合并数据集
feat = pd.concat([feat0,feat1],ignore_index=True)
```

```python
feat
```

```python
feat.to_csv('../data/feature.csv',index_label='id')
```
